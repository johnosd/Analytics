# Tipos de Variáveis

## Variável Escalar

Variáveis representadas por um único valor.

```
# Tipo Inteiro - int
idade = 27

# imprimi o valor da variável
print(idade)

# imprimi o tipo da variável
print(type(idade))
```

```
# Tipo Float - float
salario = 3478.24

# imprimi o valor da variável
print(salario)

# imprimi o tipo da variável
print(type(salario))
```

```
# Tipo booleano - bool
luz_esta_acessa = False

# imprimi o valor da variável
print(luz_esta_acessa)

# imprimi o tipo da variável
print(type(luz_esta_acessa))
```

```
# Tipo None
sem_resposta = None

# imprimi o valor da variável
print(sem_resposta)

# imprimi o tipo da variável
print(type(sem_resposta))
```

## Variável Não-Escalar

Variáveis representadas por estruturas que podem comportar quantos valores quisermos.

```
# Tipo Texto - str
nome = 'João Nogueira'

# imprimi o valor da variável
print(nome)

# imprimi o tipo da variável
print(type(nome))
```

```
# Tipo Lista - list
lista = ['Labdata', 'FIA', 'Python', 'Machine Learning']

# imprimi o valor da variável
print(lista)

# imprimi o tipo da variável
print(type(lista))
```

```
# Tipo Tupla - tuple
tupla = ('Labdata', 'FIA', 'Python', 'Machine Learning')

# imprimi o valor da variável
print(tupla)

# imprimi o tipo da variável
print(type(tupla))
```

```
# Tipo Dicionário - dict
dicionario = {'nome': 'João Paulo', 'idade': 29, 'assuntos': ['python', 'machine learning']}

# imprimi o valor da variável
print(dicionario)

# imprimi o tipo da variável
print(type(dicionario))
```

# Indexação e Fatiamento

Qualquer variável não-escalar também é chamada de objeto iterável. Isso se dá porque podemos percorré-la selecionando os elementos internos desejados.

```
lista = ['Labdata', 'FIA', 'Python', 'Machine Learning']
lista
```

### Indexação

```
# selecionando o primeiro elemento de uma lista
lista[0]
```

```
# selecionando o último elemento de uma lista
lista[-1]
```

### Fatiamento

```
# selecionando todos os elementos da lista entre as posições 1 e 2
lista[1:3]
```

```
# selecionando todos os elementos da posição 2 até o final
lista[2:]
```

# Estrutura de Controle de Fluxo

## if-else

```
idade = 27

# os dois pontos no final da linha indica que estamos iniciando um novo bloco de código
if idade < 10:
    print("Criança")
else:
    if idade < 18:
        print("Adolescente")
    else:
        print("Adulto")
```

```
a = 200
b = 33
if b > a:
    print("b é maior que a")
elif a == b:
    print("a e b são iguais")
else:
    print("a é maior que b")
```

# Estruturas de Repetição

## for loop

```
# iterando pelo elemento dentro de uma lista
produtos = ['ipad', 'celular', 'notebook', 'tv']

for produto in produtos:
    print(produto)
```

```
# iterando pelo elemento índice da lista
produtos = ['ipad', 'celular', 'notebook', 'tv']

for indice in range(len(produtos)):
    print(produtos[indice])
```

```
# somando todos os número de 0 a 99
soma = 0
for numero in range(100):
    soma = soma + numero
print(soma)
```

```
# somando todos os números de 1 até 100
soma = 0
for numero in range(1, 101):
    soma = soma + numero
print(soma)
```
