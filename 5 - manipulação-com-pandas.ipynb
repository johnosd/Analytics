# Filtrando Linhas

## Primeira Forma

Fatiamento

```
# condição simples
df[df['qualidade_cozinha'] == "Ex"]
```

```
# duas condições separadas pelo operador lógico E (&)
df[(df['qualidade_cozinha'] == "Ex") & (df['qtde_banheiros']  > 1)]
```

```
# duas condições separadas pelo operador lógico OU (|)
df[(df['qualidade_cozinha'] == "Ex") | (df['qtde_banheiros']  > 1)]
```

```
# duas condições
df[(df['qualidade_cozinha'].isin(['Ex', 'Gd'])) & (df['qtde_banheiros'] == 1)]
```

## Segunda Forma

Usando o comando `query`

```
# condição simples
df.query('qualidade_cozinha == "Ex"')
```

```
# duas condições separadas pelo operador lógica E (&)
df.query('qualidade_cozinha == "Ex" & qtde_banheiros > 1')
```

```
# duas condições separadas pelo operador lógico OU (|)
df.query('qualidade_cozinha == "Ex" | qtde_banheiros > 1')
```

```
# duas condições
df.query('qualidade_cozinha in ["Ex", "Gd"] & qtde_banheiros == 1')
```

# Criando uma Nova Coluna

## Primeira Forma
 
 Criando a coluna diretamento na tabela

```
# criando a variável ratio
df['ratio'] = df['tam_sala_estar'] / df['tam_terreno']
```

```
df[['tam_sala_estar', 'tam_terreno', 'ratio']].head()
```

```
# removendo a coluna criada
df = df.drop('ratio', axis=1)
```

## Segunda Forma

Usando o comando `assign`

```
# o comando assign não cria a variável nova diretamente na tabela
# esse comportamento é bom para validação antes de dar o ok da nova variável
df.assign(ratio = df['tam_sala_estar'] / df['tam_terreno'])
```

```
# podemos também concatenar comandos
df.assign(ratio = df['tam_sala_estar'] / df['tam_terreno']).filter(['tam_sala_estar', 'tam_terreno', 'ratio']).head()
```

```
# podemos concatenar o código colocando todo mundo dentro de parênteses
# dessa forma a leitura do código fica mais limpa e clara a leitura
(
    df
    .assign(ratio = df['tam_sala_estar'] / df['tam_terreno'])
    .filter(['tam_sala_estar', 'tam_terreno', 'ratio'])
    .head()
)
```

```
# se por acaso deseja salvar o resultado da manipulação acima,
# basta atribuir toda a operação a uma nova variável
new_df = (
    df
    .assign(ratio = df['tam_sala_estar'] / df['tam_terreno'])
    .filter(['tam_sala_estar', 'tam_terreno', 'ratio'])
    .head()
)
```

# Agregando Colunas

Comando `agg`

# Group By

```
# agrupamento por 1 coluna
df.groupby('qualidade_cozinha').agg(preco_medio = ('preco', 'mean'))
```

```
# resetando o index da tabela agrupada
df.groupby('qualidade_cozinha').agg(preco_medio = ('preco', 'mean')).reset_index()
```

```
# agrupando por duas colunas
df.groupby(['qualidade_cozinha', 'ar_condicionado']).agg(preco_medio = ('preco', 'mean')).reset_index()
```

```
# agrupando por duas colunas e criando
# duas variáveis agregadas
df.groupby(['qualidade_cozinha', 'ar_condicionado']).agg(preco_medio = ('preco', 'mean'), preco_max = ('preco', 'max')).reset_index()
```

Perceba que o comando anterior já começou a ficar bem grande! Aqui já é uma boa opção utilizamos escrevermos o comando de forma indentada dentro de chaves

```
(
    df
    .groupby(['qualidade_cozinha', 'ar_condicionado'])
    .agg(preco_medio = ('preco', 'mean'),
         preco_min = ('preco', 'min'),
         preco_max = ('preco', 'max'))
    .sort_values(by='preco_medio', ascending=False)
    .reset_index()
)
```

```
# podemos pegar o resultado anterior e apresentá-lo em um formato de 
# tabela dinâmica do excel com o comando unstack(). Para isso, não
# devemos executar o comando reset_index()
(
    df
    .groupby(['qualidade_cozinha', 'ar_condicionado'])
    .agg(preco_medio = ('preco', 'mean'),
         preco_min = ('preco', 'min'),
         preco_max = ('preco', 'max'))
    .sort_values(by='preco_medio', ascending=False)
    .unstack()
)
```

O comando `unstack()` pega o segundo índice da tabela, que no caso acima era o `ar_condicionado` e joga para a coluna.

# Join

## Left Join

Similar ao procv do excel. Dá o match nos campos especificados de duas tabelas que queremos juntar, mas mantendo todos os valores de campos da tabela da esquerda.

```
funcionarios = pd.DataFrame({
    'funcionario': ['José', 'Maria', 'Chico', 'Antônia', 'Edgar', 'Joana'],
    'cargo_id':[3, 1, 3, 2, 2, 4]
})
funcionarios
```

```
cargos = pd.DataFrame({
    'cargo_id': [1, 2, 3],
    'cargo': ['Arquiteto Java', 'Desenvolvedor Java Sr.', 'Desenvolvedor Java Jr.']
})
cargos
```

```
funcionarios.merge(cargos, left_on='cargo_id', right_on='cargo_id', how='left')
```

## Inner Join

Traz apenas os valores que existem tanto na tabela da esquerda quanto na da direita.

```
funcionarios.merge(cargos, left_on='cargo_id', right_on='cargo_id', how='inner')
```

# Concatenação

```
funcionarios1 = pd.DataFrame({
    'funcionario': ['José', 'Maria', 'Chico'],
    'cargo_id':[3, 1, 3]
})
funcionarios1
```

```
funcionarios2 = pd.DataFrame({
    'funcionario': ['Antônia', 'Edgar', 'Joana'],
    'cargo_id':[2, 2, 4]
})
funcionarios2
```

```
# concatenando as bases uma em cima da outra (as duas bases tem que ter as mesmas colunas)
pd.concat([funcionarios1, funcionarios2])
```

```
# concatenando as bases uma do lado da outra (as duas bases tem que ter a mesma quantidade de linhas)
pd.concat([funcionarios1, funcionarios2], axis=1)
```
